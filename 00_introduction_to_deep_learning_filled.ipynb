#### Neural networks with Keras

Now we will build the model shown above using Keras library.

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(5, input_dim=3, activation="tanh"))
```

```python
model.summary()
```

Let us apply the current model to some random data instance of appropriate size.

```python
import numpy as np
```

```python
x = np.random.rand(1, 3)
y = model(x)
print("x =", x)
print("y =", y)
```

Now we are half-way to building our model. Let us add the output layer so that our model will return two numbers.

```python
model.add(Dense(2))
```

```python
model.summary()
```

```python
x = np.random.rand(1, 3)
y = model(x)
print("x =", x)
print("y =", y)
```

Instead of directly applying a model to the input data it is more usual to use `predict` method of the tensorflow object we call model.

```python
y = model.predict(x)
print("x =", x)
print("y =", y)
```

#### Activation functions

In this example we used hyperbolic tangent as a nonlinearity that is applied on every node in hidden layer. However, there are many to choose from. Yet, current research show that this choice makes little difference. All it matters is that it is nonlinear.

a_g.PNG

By far the most widely used activation function is called rectified linear unit or "relu" for short and is simply defined as max(0, x) and is therefore computationally very cheap to compute. 

relu.png

There is one guideline though, to circumvent vanishing gradient problem the activation function is advised to be unbounded (at least on one side). However, more modern research shows that more intense (longer) training oftentimes resolves these issues.

### Deep neural network

Neural network is deep if it has more than one hidden layer. This means that nonlinearity is applied multiple times which means that the model can capture more complicated functions.

deep.PNG

Let us implement deep neural network shown on the image above in Keras.

```python
model = Sequential()
model.add(Dense(4, input_dim=3, activation="relu"))
model.add(Dense(4, activation="relu"))
model.add(Dense(1))
```

```python
model.summary()
```

This time we have used the rectified linear unit for the activation function. This is the most common choice in Deep learning.

```python
x = np.random.rand(1, 3)
y = model.predict(x)
print("x =", x)
print("y =", y)
```

There are "deeper" reasons why deep neural networks capture wider range of functions. It has to do with symmetry of neural networks. It turns out that capacity of the neural network grows as a power of width, while it grows exponentially with depth (see eg. [this paper](https://arxiv.org/pdf/1402.1869.pdf)).

**However, in practice it is quite unusual to see more than 3 hidden layers because it turns out that a network with 1 or 2 layers works better or equally well than eg. 10 layer network.**

### Deep learning
Official definition: Deep learning is the study of machine learning
models composed of multiple layers of functions that progressively
extract higher level features from the raw input.
However, this idea existed also 1950-2010 when success of deep
learning was very limited.

What really made it so superior in the last 10 years:
* gradient based training (on GPU)
* availability of large quantity of data
* appropriate cost functions
* new regularizations
* new representation mappings (eg. embeddings)
* new network architectures
